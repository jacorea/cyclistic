## 4. Dataset Summary

The client provided KPMG with 4 datasets:

* Customer Demographic
* Customer Addresses
* Transactions data in the past 3 months
* New Customers

Customer Addresses has their address information such as address, postcode, state, country and property valuation. Transactions table has their transactions data such as product information, transaction date, id, and product first sold date. The last table New Customers has new customer's demographic data which is almost the same as Customer Demographic table but with new customers.
